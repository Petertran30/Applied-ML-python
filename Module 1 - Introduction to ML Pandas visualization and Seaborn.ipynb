You can also choose the plot kind by using the `DataFrame.plot.kind` methods instead of providing the `kind` keyword argument.

`kind` :
- `'line'` : line plot (default)
- `'bar'` : vertical bar plot
- `'barh'` : horizontal bar plot
- `'hist'` : histogram
- `'box'` : boxplot
- `'kde'` : Kernel Density Estimation plot
- `'density'` : same as 'kde'
- `'area'` : area plot
- `'pie'` : pie plot
- `'scatter'` : scatter plot
- `'hexbin'` : hexbin plot

# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0,10,1000))
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000))
```

```python
v1.describe()
```

```python
v2.describe()
```

```python
plt.figure()
plt.hist(v1, alpha = 0.7, bins=np.arange(-50,150,5), label = "v1") # bins is set to imclude the wider range among two series
plt.hist(v2, alpha = 0.7, bins=np.arange(-50,150,5), label = "v2")
plt.legend(frameon = False)
```

```python
# plot a kernel density estimation over a stacked barchart
plt.figure()
plt.hist([v1,v2], histtype = "barstacked", density = True);

v3 = np.concatenate((v1,v2))
sns.kdeplot(v3)
```

```python
plt.figure()
# we can pass keyword arguments for each individual component of the plot
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'});
```

```python
sns.jointplot(v1, v2, alpha = 0.4);
```

```python
grid = sns.jointplot(v1, v2, alpha = 0.4);
grid.ax_joint.set_aspect('equal')
```

```python
sns.jointplot?
```

```python
grid = sns.jointplot(v1, v2, alpha = 0.4, xlim = (-60, 100))
grid.ax_joint.set_aspect("equal")
```

```python
plt.figure()
sns.jointplot(v1,v2, kind = "hex")
```

```python
# set the seaborn style for all the following plots
sns.set_style('white')

sns.jointplot(v1, v2, kind='kde', space=0); # space = 0 to tell seaborn that we donot want space between subplots
```
